7.Replace the numbers in the column published_month as names of the months i,e., 1 as 'Jan', 2 as 'Feb'
and so on.....?

8. Find the number of videos published each month and arrange the months in a decreasing order based
on the video count.?

```python
#useing pivot_table creating video count by decreasing order
pd.pivot_table(data,index=['published_month'],values=['video_id'],aggfunc='count').sort_values('video_id',ascending=False).head(10)
#oct month has most videos and Apr month has less video count
```

10. Find the top10 channel names having the highest number of videos in the dataset and the bottom10

11. Find the title of the video which has the maximum number of likes and the title of the video having
minimum likes and write your inferences.?

```python
#Video title with maximum number of likes.useing pivot_table
pd.pivot_table(data,index=['title'],values=['likes'],aggfunc='max').sort_values('likes',ascending=False).head(1)
#BTS () 'Dynamite' Official MV title has most  number of likes
```

```python
#Video title with minimum number of likes.useing pivot_table
pd.pivot_table(data,index=['title'],values=['likes'],aggfunc='min').sort_values('likes',ascending=True).head(1)
#Leading the Charge | Circle K has low likes
```

12. Find the title of the video which has the maximum number of dislikes and the title of the video having

13. Does the number of views have any effect on how many people disliked the video? Support your
answer with a metric and a plot.?

```python
#finding correalation on 'view_count','dislikes'
data.corr()
```

```python
#showing correalation in seaborn heatmap
sns.heatmap(df[['view_count','dislikes']].corr(), annot = True, cmap = 'magma')
plt.show()
```

```python
#showing relationship between 'view_count','dislikes' useing scatter plot
x=data['view_count']
y=data['dislikes']
plt.scatter(x,y)

#adding tittle 
plt.title('difference between view_count,dislikes')

#adding labels 
plt.xlabel('view_count')
plt.ylabel('dislikes')
plt.show()
```

```python
#From the above correaltion and scatter plot we can say that there is a positive relationship between view_count and dislikes
#As view_count increases dislikes also increases
# There is perfect linear relationship between view_count and dislikes
```
